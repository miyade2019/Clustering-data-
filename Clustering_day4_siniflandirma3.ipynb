```python
x=df[['Price','Mileage','Cylinder']]
```

```python
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
```

```python
x.head()
```

```python
model=KMeans(4)
```

```python
model=KMeans(7)# 7 gurpa ayırıldı
```

```python
model=model.fit(x)
```

```python
tahmin=model.predict(x)
```

```python
tahmin
```

```python
x['cluster']=tahmin #yeni gurp açıyoruz
```

```python
x.head()
```

```python
silhouette_score(x,tahmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x.Mileage,x.Price,c=tahmin)
plt.colorbar()
plt.xlabel('km')
plt.ylabel('fiyat')
```

```python
import numpy as np

from matplotlib.colors import ListedColormap

renkler = ['red', 'darkred', 'indianred', 'firebrick']

cmap = ListedColormap(renkler)

plt.figure(figsize=(10,6))

plt.scatter(x.Mileage, x.Price, c=tahmin, cmap=cmap)

plt.colorbar();

plt.xlabel("Km")

plt.ylabel("Fiyat")
```

```python
#wcss= within cluster sum of squares

wcss=[]

ss=[]

for i in range(2,10):

    model=KMeans(i)

    model=model.fit(x)

    tahmin=model.predict(x)

    ss1=silhouette_score(x,tahmin)

    ss.append(ss1)

    print(ss1)

    wcss.append(model.inertia_)
```

```python
plt.plot(range(2,10),wcss)
```

```python
sns.lineplot(x=range(2,10),y=wcss,marker='*')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
df=pd.read_csv('Billionaires.csv')


# Get the value counts and sort them
ydf = df['country'].value_counts().sort_values(ascending=False)

# Convert the Series to a DataFrame for easier plotting
ydf = ydf.reset_index()
ydf.columns = ['country', 'count']

# Plot the barplot
plt.figure(figsize=(20, 10))
sns.barplot(x='count', y='country', data=ydf, palette='viridis')
plt.xlabel('Counts')
plt.ylabel('Country')
plt.title('Country Counts')
plt.show()
```
